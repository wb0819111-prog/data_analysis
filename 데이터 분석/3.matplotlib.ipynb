#**1. Matplotlib**
[Matplotlib](https://matplotlib.org/)은 파이썬에서 데이터를 시각화하는 데 널리 사용되는 강력한 라이브러리입니다. 다양한 그래프와 차트를 그릴 수 있으며, 선 그래프, 막대 그래프, 히스토그램, 산점도 등 기본적인 그래프부터 복잡한 3D 플롯까지 지원합니다. 사용법이 비교적 간단하고, 커스터마이징이 가능하여 데이터의 패턴과 트렌드를 효과적으로 표현할 수 있습니다. 또한, NumPy와 Pandas와 같은 데이터 분석 라이브러리와 잘 통합되어 데이터 과학, 머신러닝, 통계 등 다양한 분야에서 활용됩니다. Matplotlib의 기본 모듈인 pyplot은 MATLAB과 유사한 인터페이스를 제공해 초보자도 쉽게 사용할 수 있도록 설계되었습니다.

###1. 기본 구조
 - Figure: 전체 그림 (도화지)
 - Axes: 하나의 서브 그래프 (보통 x축, y축 포함)
 - Axis: x축과 y축 자체

```python
# Figur: 도화지의 크기,색상,도화지에 그리는 글자의 폰트 이런것 설정
# Axes: 그래프의 모양,크기 바꾸는 것
```

```python
import matplotlib.pyplot as plt  #matplotlib 그래프그리는 모듈
```

```python
# 리스트의 값들은 y값들이며, x값은 자동으로 만들어짐#############################
plt.plot([1, 2, 3, 4]) # 리스트의 값들은 y값들이며, x값은 자동으로 만들어줌
plt.show()
```

```python
 #           x축             y축
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.show()
```

```python
import numpy as np
```

```python
data = np.arange(1, 100) # 1~99까지 y축
plt.plot(data)
plt.show()
```

```python
data1 = np.arange(1, 50)
plt.plot(data1)
data2 = np.arange(50, 100)
plt.plot(data2)  #추가하면 누적이 된다 ########################x값은 자동으로
plt.show()
```

```python
# 선 그래프	  |  plt.plot()	  |  연속적인 데이터 추세
# 막대 그래프	|  plt.bar()	  |  범주형 데이터 비교
# 수평 막대	  |  plt.barh()	  |  막대를 가로로
# 산점도	    |  plt.scatter()|  두 변수의 관계 시각화
# 히스토그램  |	plt.hist()	  |  분포 시각화
# 파이 차트	  |  plt.pie()	  |   비율 표현
```

```python
plt.figure(figsize=(4, 5)) # inch  #fugure도화지 설정/ figsize=(4,5) = 가로4인치, 세로5인치 그라는 의미################
plt.plot([1, 2, 3], [1, 2, 3])
plt.title('제목', fontsize=30)  #그래프 제목
plt.xlabel('X축', fontsize=20)  #x축 이름
plt.ylabel('Y축', fontsize=20, rotation=0)  #y축 이름
plt.show()
# 한글이 들어가 있어서 따로 설치 필요
```

```python
plt.figure(figsize=(15, 10))
plt.title('마커설정', fontsize=30)
plt.plot(np.arange(10), np.arange(10), color='deepskyblue', marker='o', linestyle='-.')
plt.plot(np.arange(10), np.arange(10)*2, color='deeppink', marker='v', linestyle='--')
plt.plot(np.arange(10), np.arange(10)*3, color='gold', marker='*', ms=10, linestyle='')
# np.arange(10)=> 0~9까지 x,y/marker='o' => 동그라미 모양/linestyle='-.선모양/ms=10 => marker모양의 크기


# 범례
# 기본 위치는 왼쪽 상단
plt.legend(['10', '10*2', '10*3'], fontsize=15, loc='lower right', ncol=3)
# loc='lower right'=>location위치조정, 기본위치 우측상단/ncol=3=>가로3개로 되어있는데 세로 3개 나열해줌

plt.xlim(0, 12) #x의 범위 보여줌
plt.ylim(0, 30) #y의 범위 보여줌

plt.xticks(rotation=30) #xticks 글자를 의미/(rotation=30) 글자를 틀어줌
plt.yticks(rotation=30)

plt.savefig('my_plot.png')  # 그래프를 이미지로 저장(avefig)/ show전에 한다 왜? show를 하고 나면 세팅값들이 다 날라간다
#다운로드 받아서 할 수 있다

plt.show()
```

```python
x = ['파이썬', '웹개발', '데이터분석', '머신러닝', '딥러닝', '컴퓨터비전']
y = [95, 70, 75, 60, 50, 30]
plt.figure(figsize=(8, 5))
plt.title('AI 성적표', fontsize=25)
plt.ylabel('수강생 점수')
plt.bar(x, y, alpha=0.5, color='deeppink')  #alpha=0.5 =>투명도
plt.show()
```

```python
x = ['파이썬', '웹개발', '데이터분석', '머신러닝', '딥러닝', '컴퓨터비전']
y = [95, 70, 75, 60, 50, 30]
plt.figure(figsize=(8, 5))
plt.title('AI 성적표', fontsize=25)
plt.xlabel('수강생 점수')
plt.barh(x, y, alpha=0.5, color='deepskyblue') #barh => 옆으로 눕힘
plt.show()
```

```python
import pandas as pd

x = ['파이썬', '웹개발', '데이터분석', '머신러닝', '딥러닝', '컴퓨터비전']
y = [95, 70, 75, 60, 50, 30]
df = pd.DataFrame({"과목":x, "점수":y})
df
```

```python
fig, ax = plt.subplots(figsize=(5, 4))
ax.bar(df["과목"], df["점수"])
ax.set_xlabel("과목", fontsize=12)
ax.set_ylabel("점수", fontsize=12)
ax.set_title("코리아 학생 점수", fontsize=20)
ax.set_xticks(range(len(df.index)))  #xticks범위
#xticks=> x의 컬럼값들 range(적용을해서)/df의 인덱스 길이만큼으로 가져와서 그것을 레인지 범위를 해서 xticks를 세팅해줘라는 의미
plt.tight_layout() # 레이아웃이 겹치지 않도록 자동으로 정리해주는 함수
plt.show()
```
