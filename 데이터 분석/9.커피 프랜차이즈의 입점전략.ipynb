#**2. 하버사인 공식**
Haversine 공식은 구체(구면) 위의 두 점(위도와 경도로 표시됨) 사이의 최단 거리(대원 거리, great-circle distance)를 계산하는 방법입니다. 이 공식은 GPS 좌표(위도와 경도)를 활용하여 지구 표면 위의 거리 계산에 널리 사용됩니다.

```python
from haversine import haversine
```

```python
df_cross['거리'] = df_cross.apply(lambda x: haversine([x['위도'], x['경도']], [x['지점위도'], x['지점경도']], unit='m'), axis=1)
df_cross
#df_cross['거리'] -> 컬럼만듬
# df_cross의 데이터를 하나씩 꺼내서 lam
# lambda x 여기로 이동
# haversine 함수안에다가 A= [x['위도'], x['경도']] => x가 전체행이니까 전체 행에서 위도와 경도를 꺼낸다 위도경도를 하나로 묶는다 => 프랜차이점에 위도경도
# B = [x['지점위도'], x['지점경도']] => 지점위도,지점경도는 스타벅스 위도경도/ 이 두개를 묶는다
# A점(위경도)과 B점(지점위경도)의 차를 구할 수 있다
# unit='m' => 단위는 m로 한다/ km 킬로미터
# axis=1 => 행으로 계산


```

```python
# 개별 매장들과 스타벅스와의 최소 거리

# '상가업소번호' => 하나의 지점/ groupby 하나로 합침/ 상가업소번호가 group으로 하면 하나의 지점이 된다 그거를 그룹으로 묶고
# 그 중에서 가장 가까운 거리에 있는 거리를 구하면 한 지점당 가장 가까운 스타벅스를 찾게 된다 => ['거리'].min()
#.reset_index() 인덱스 다시
# 그러면 한 업체당 가장 가까운 스타벅스가 나온다






df_dis = df_cross.groupby(['상가업소번호','상호명'])['거리'].min().reset_index()
df_dis
```

```python
# 각 프렌차이즈 별 스타벅스와의 평균 거리
df_dis.groupby('상호명')['거리'].mean()
```

```python
df_dis.groupby('상가업소번호')['거리'].mean()
```

```python
# agg(): 다중 집계작업을 간단하게 해주는 함수 집계 작업 두개 이상 평균, 개수 #df_dis => 최솟값 구한 기준으로
df_dis.groupby(['상호명'])['거리'].agg(['mean','count'])
# 최솟값을 낸 상태에서 그 최솟값 중에서 상호명으로 평균가 개수를 구한다

```

```python
# 거리를 입력하면 프렌차이즈 별 스타벅스와의 평균거리와 매장개수를 출력하는 함수
def distance(x):  # 만약 100m라면
    dis = df_dis['거리'] <= x  # 100m 보다 작은 것(<=)을 df_dts['거리']를 뽑아서 dis에 저장
    return df_dis[dis].groupby('상호명')['거리'].agg(['mean', 'count']) # 그런다음 평균과 개수 가져온다

distance(100)




```

```python
from pyecharts.charts import Timeline, Grid

tl = Timeline({'width':'600px', 'height':'400px'})  # 밑에 있는 것 넓이:600, 높이:400

for i in [1000, 100, 50, 30]:
    df_d = distance(i).reset_index()
    pie1 = df_d.echart.pie(x='상호명', y='count', figsize=(600, 400),
                    radius=['20%', '60%'], label_opts={'position':'outer'},  #radius=['20%', '60%'] => 파이차트의 안쪽 반지름, 바깥쪽 반지름 퍼센트 지정 # label_opts={'position':'outer'} => 라벨을 파이 조각 바깥쪽에 표시해라
                    title='커피 프렌차이즈의 입점전략은 과연 스타벅스 옆인가?',
                    legend_opts={'pos_right':'0%', 'orient':'vertical'},  # legend_opts={'pos_right':'0%', 'orient':'vertical'} => pos_right='0%': 범례를 오른쪽 끝에 붙여서 정렬해라(가장 오른쪽) / 'orient':'vertical': 범례를 세로(수직)방향으로 나열해라
                    subtitle='{}m 이내 매장수'.format(i), init_opts={'bg_color': 'white'})   # init_opts={'bg_color': 'white'} => 그래프의 배경색을 흰색으로 설정해라
    tl.add(pie1, '{}m'.format(i)).render()  #pie1을 타임라인에 넣어주고  render()를 하면 위에랑 아래랑 연결이 된다
    #'{}m' => 1000, 100, 50, 30 이거를 각각 그래프 4개를 만들었다 그거를 for문 돌려서 타임라인에 따라서 4분류가 각각 다르게 보이게끔 만든 것
    # 파이 4개
IPython.display.HTML(filename='render.html')


```
